```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
prices = yf.download(['^BVSP','PETR4.SA','VALE3.SA', 'BBAS3.SA'],
                     group_by="column").dropna()
```

```python
# Plot Brazilian Index and Stocks

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(go.Scatter(x = prices.index, y = prices[('Adj Close',    '^BVSP')],
                         name = 'IBOVEPSA'),
                        secondary_y=False)
fig.add_trace(
  go.Scatter(x = prices.index, y = prices[('Adj Close', 'BBAS3.SA')],
             name = 'BBAS3'),
    secondary_y=True
)
fig.add_trace(
  go.Scatter(x = prices.index, y = prices[('Adj Close', 'PETR4.SA')],name = 'PETR4'),
    secondary_y=True
)

fig.add_trace(
  go.Scatter(x = prices.index, y = prices[('Adj Close', 'VALE3.SA')],
             name = 'VALE'),
    secondary_y=True
)
fig.update_layout(
    template=  "plotly_white",
    title_text="IBOVESPA INDEX and Main Brazilian Stocks"
)
fig.update_yaxes(title_text="Ibovespa Index", secondary_y=False)
fig.update_yaxes(title_text="Stocks Prices", secondary_y=True)
fig.show()

```
